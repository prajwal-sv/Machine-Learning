```python
import seaborn as sns
tips = sns.load_dataset("tips")
print(tips.head())

```

```python
#scatter ploT
sns.scatterplot(x='total_bill', y='tip',data=tips)
```

```python
#categorical scatter plot
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
#bar plot
sns.barplot(x='day', y='total_bill', data=tips, palette='Blues')
```

```python
#box plot
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
```

```python
#violin plot
sns.violinplot(x='day', y='total_bill', data=tips, palette='Pastel1')
```

```python
#histogram
sns.histplot(tips['total_bill'], bins=20, kde=True, color='purple', edgecolor='black', linewidth=1.5)
```

```python
#kde plot
sns.kdeplot(tips['total_bill'], color='orange', fill=True, alpha=0.5, linewidth=2)
```

```python
#pairplot
sns.pairplot(tips, hue='day', palette='Set1')
```

```python
# Heatmap (numeric columns only)

import seaborn as sns
tips = sns.load_dataset("tips")
print(tips.head())
numeric_tips = tips.select_dtypes(include='number')
sns.heatmap(numeric_tips.corr(), annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black', fmt='.2f', cbar_kws={'label': 'Correlation Coefficient'})
```
