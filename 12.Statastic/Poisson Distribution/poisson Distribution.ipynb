```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 1. Define 'mu' (often called lambda Î»)
# This is the average number of events you expect in the interval.
average_events = 3 

# 2. Create a range of possible outcomes (x-axis)
# We want to see the probability for 0, 1, 2... up to 10 events.
x = np.arange(0, 11)
```

```python
# 3. Calculate the Probability Mass Function (PMF)
# This tells us the exact probability for each value in 'x'.
probabilities = poisson.pmf(x, average_events)
```

```python
# --- Visualization ---
plt.bar(x, probabilities, color='skyblue', edgecolor='black')
plt.title(f'Poisson Distribution (Average = {average_events})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.xticks(x)
plt.show()
```
