```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Parameters
population_size = 100000
sample_size = 40      # "n" from our formula
num_samples = 10000   # How many times we take a sample
```

```python
population = np.random.uniform(0, 100, population_size)
```

```python
# 3. Take multiple samples and calculate their means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))
```

```python
# 4. Visualization
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=50, color='skyblue', edgecolor='black', density=True)
plt.title(f'Distribution of {num_samples} Sample Means (n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
