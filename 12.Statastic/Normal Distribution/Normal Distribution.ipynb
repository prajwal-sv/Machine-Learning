```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Set the parameters
mu = 175  # Mean (average height in cm)
sigma = 7 # Standard Deviation (spread)

# 2. Generate random data points (1000 men)
data = np.random.normal(mu, sigma, 1000)

# 3. Create the plot
count, bins, ignored = plt.hist(data, 30, density=True, alpha=0.6, color='skyblue')

# 4. Plot the theoretical probability density function (the "Bell Curve")
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), color='darkblue', linewidth=2)

plt.title("Normal Distribution of Heights (cm)")
plt.xlabel("Height")
plt.ylabel("Probability Density")
plt.show()
```
