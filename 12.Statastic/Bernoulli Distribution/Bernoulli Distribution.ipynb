The Bernoulli Distribution is the simplest building block in all of statistics. If the Binomial distribution is "flipping a coin 20 times," the Bernoulli distribution is flipping that coin exactly once.It models a single trial with only two possible outcomes: Success (1) or Failure (0).

The Logic of Bernoulli$p$: The probability of success (e.g., 0.7 chance a user clicks).$q$: The probability of failure, which is always $1 - p$ (e.g., 0.3 chance they don't).

Data Science Use Case: This is how we represent a single row in a dataset for "Churn" (Left vs. Stayed) or "Spam" (Spam vs. Not Spam).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
# 'p' is the probability of success.
# Let's say there is a 70% chance a customer makes a purchase.
p = 0.7 

# The possible outcomes for a Bernoulli trial are ALWAYS only 0 and 1.
# 0 = Failure (No purchase), 1 = Success (Purchase)
outcomes = [0, 1]
```

```python
# PMF (Probability Mass Function):
# This tells us the height of the bars for 0 and 1.
# For Bernoulli: PMF(1) = p, and PMF(0) = 1 - p
pmf_values = bernoulli.pmf(outcomes, p)
```

```python
# In Data Science, we often "simulate" data. 
# Let's simulate 1,000 customers to see the distribution in action.
# 'size' determines how many independent trials we perform.
simulated_data = bernoulli.rvs(p, size=1000)
```

```python
plt.figure(figsize=(10, 5))

# Plotting the theoretical Probability (what 'should' happen)
plt.bar(outcomes, pmf_values, color=['red', 'green'], alpha=0.6, width=0.4)

# Adding labels for clarity
plt.title(f'Bernoulli Distribution (p={p})')
plt.xlabel('Outcome (0 = Failure, 1 = Success)')
plt.ylabel('Probability')
plt.xticks(outcomes) # Only show 0 and 1 on the X-axis

# Add text labels on top of the bars to show the values
for i, val in enumerate(pmf_values):
    plt.text(i, val + 0.02, f'{val*100:.0f}%', ha='center', fontweight='bold')

plt.show()
```

```python
# The Mean of a Bernoulli distribution is always equal to 'p'.
mean = bernoulli.mean(p)

# The Variance is p * (1 - p). 
# This represents how "uncertain" the outcome is.
# Variance is highest when p = 0.5 (like a fair coin flip).
var = bernoulli.var(p)

print(f"Expected Mean (p): {mean}")
print(f"Variance (p * q): {var:.4f}")
print(f"Simulated Success Rate: {np.mean(simulated_data):.4f}")
```
