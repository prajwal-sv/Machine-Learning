```python
import matplotlib.pyplot as plt
#matlotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
#pyplot is a collection of command style functions that make matplotlib work like MATLAB.

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
#create line plot
plt.plot(x, y)

plt.title("Simple Line Plot")
plt.xlabel("X-axis")    
plt.ylabel("Y-axis")
plt.grid(True)  # Add grid lines for better readability
```

```python
#costomize line plot
plt.plot(x, y, color='green', linestyle='--', marker='o', markersize=8, linewidth=2)
```

```python
#bar plot
plt.bar(x, y, color='blue', width=0.4)
```

```python
#pie chart
labales = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 45, 10, 5]
plt.pie(sizes, labels=labales, autopct='%1.1f%%', startangle=140)
```

```python
# histogram
a = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(a, color='purple', edgecolor='black')
```
